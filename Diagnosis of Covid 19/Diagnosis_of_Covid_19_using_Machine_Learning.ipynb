```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")


URL ='https://drive.google.com/file/d/1NvCJczRMASLsUEI-_A8X-29ryNNufP8P/view?usp=sharing'
data = 'https://drive.google.com/uc?export=download&id='+URL.split('/')[-2]
df = pd.read_csv(data,delimiter=',')
df
```

```python
df.corr().abs() #kolerasi
```

##Memilih fitur yang akan di ukur 

```python
X = df.drop(columns=['Terinfeksi-COVID'])
y = df['Terinfeksi-COVID']
```

##Splitting Dataset untuk Training dan Testing (Ratio 80:20)

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

##Build Model Multinomial NB, Categorical NB, & Decision Tree Classifier

```python
from sklearn.naive_bayes import MultinomialNB, CategoricalNB
from sklearn.tree import DecisionTreeClassifier
#multimonial > fitur di encode, categorical > fitur yg diskit(udah berbentuk angka)
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, plot_confusion_matrix
import matplotlib.pyplot as plt

model1 = MultinomialNB()
model2 = CategoricalNB()
model3 = DecisionTreeClassifier()
```

## Evaluasi Model 1 - Multibinomial NB

```python
print('Hasil modelling model 1 MultinomialNB :')

model1.fit(X_train, y_train)
prediction1 = model1.predict(X_test)

print('Akurasi model 1 MultinomialNB :')
print(accuracy_score(y_test, prediction1))

print('Confusion Matrix :')
plot_confusion_matrix(model1, X_test, y_test)

print('Classification report :')
print(classification_report(y_test, prediction1))
```

##Evaluasi Model 2-CategoricalNB

```python
print('Hasil modelling model 2 CategoricalNB :')

model2.fit(X_train, y_train)
prediction2 = model2.predict(X_test)

print('Akurasi model 1 CategoricalNB :')
print(accuracy_score(y_test, prediction2))

print('Confusion Matrix :')
plot_confusion_matrix(model2, X_test, y_test)

print('Classification report :')
print(classification_report(y_test, prediction2))
```

##Evaluasi Model 3- Decision Tree Classifier

```python
print('Hasil modelling model 3 DecisionTreeClassifier :')

model3.fit(X_train, y_train)
prediction3 = model3.predict(X_test)

print('Akurasi model 1 CategoricalNB :')
print(accuracy_score(y_test, prediction3))

print('Confusion Matrix :')
plot_confusion_matrix(model3, X_test, y_test)

print('Classification report :')
print(classification_report(y_test, prediction3))
```
